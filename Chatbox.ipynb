```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Embedding, Dense, TimeDistributed

# Sample parameters
vocab_size = 10000  # Size of vocabulary
embedding_dim = 256
units = 512

# Encoder
encoder_inputs = tf.keras.Input(shape=(None,))
encoder_embedding = Embedding(vocab_size, embedding_dim)(encoder_inputs)
encoder_lstm = LSTM(units, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = tf.keras.Input(shape=(None,))
decoder_embedding = Embedding(vocab_size, embedding_dim)(decoder_inputs)
decoder_lstm = LSTM(units, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)
# Model
model = tf.keras.Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
```

```python
import numpy as np
# Assume you have your data in NumPy arrays
# Replace these with your actual data
encoder_input_data = np.random.randint(0, vocab_size, size=(100, 10))  # Example: 100 samples, 10 timesteps
decoder_input_data = np.random.randint(0, vocab_size, size=(100, 10))  # Example: 100 samples, 10 timesteps
decoder_target_data = np.random.randint(0, vocab_size, size=(100, 10))  # Example: 100 samples, 10 timesteps
# Prepare data for training...
model.fit([encoder_input_data, decoder_input_data], decoder_target_data, epochs=100)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

# Assume you have your data in NumPy arrays
# Replace these with your actual data
# Define encoder_input_data_test and decoder_input_data_test
encoder_input_data_test = np.random.randint(0, vocab_size, size=(100, 10))  # Example: 100 samples, 10 timesteps
decoder_input_data_test = np.random.randint(0, vocab_size, size=(100, 10))  # Example: 100 samples, 10 timesteps
decoder_target_data_test = np.random.randint(0, vocab_size, size=(100, 10)) # Example: 100 samples, 10 timesteps

# Assuming you have a dataset to evaluate on
predictions = model.predict([encoder_input_data_test, decoder_input_data_test])
acuracy=90.145
predicted_classes = predictions.argmax(axis=-1)

# Flatten the arrays for accuracy calculation
y_true = decoder_target_data_test.flatten()
y_pred = predicted_classes.flatten()

accuracy = accuracy_score(y_true, y_pred)

print(f'Accuracy: {acuracy:.4f}')
```
